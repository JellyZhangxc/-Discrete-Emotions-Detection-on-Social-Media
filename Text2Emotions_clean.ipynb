### 1 Load the data and extract a subset as input

Text2emotion processes the data and yield the emotion of the sentence from these 5 emotions (Happy, Angry, Sad, Surprise, and Fear)

```python
import pandas as pd
```

```python
# Replace 'your_file.csv' with the path to your CSV file
file_path = '.\Source\GoEmotions_1000_Clean.csv'
# Load the data
raw_df = pd.read_csv(file_path)
sub_df=raw_df[raw_df['primary_emotion'].isin(['joy', 'surprise', 'anger', 'fear', 'sadness', 'neutral'] )]
```

```python
display(sub_df)
```

### 2  Loop the 600 samples to get emotion

```python

# !pip install text2emotion
# !pip uninstall emoji
# !pip install emoji==1.7
# https://github.com/aman2656/text2emotion-library/issues/10
```

```python
import text2emotion as te
```

```python
#igore trust 
emotions_list = ['Happy', 'Angry', 'Surprise', 'Sad', 'Fear']  # 'neutral',    
em_df = pd.DataFrame(columns=emotions_list+["id"])
display(em_df)
```

```python
len(sub_df)
```

```python
for  index, row in sub_df.iterrows():
    sentence=row["clean_text"]
    scores =te.get_emotion(sentence)  
    # print(scores)  #e.g. {'Happy': 0.5, 'Angry': 0.0, 'Surprise': 0.0, 'Sad': 0.5, 'Fear': 0.0}
    # Create a row with the scores for the specified emotions and 0 for the rest
    em = {emotion: scores.get(emotion, 0) for emotion in emotions_list}
    em['id'] = row['id']
    # Convert the row dictionary to a DataFrame and concatenate with the existing DataFrame
    em_df = pd.concat([em_df, pd.DataFrame([em])], ignore_index=True)
```

```python
len(em_df)
```

```python
#if  all 0 for ''Happy', 'Angry', 'Surprise', 'Sad', 'Fear' , then neutral
em_df['neutral'] = em_df.apply(lambda row: 1 if all(row[:5] == 0) else 0, axis=1)
em_df=em_df[['Happy', 'Angry', 'Surprise', 'Sad', 'Fear','neutral','id']]
#Rename columns: align  emotions from text2emotion with   primary_emotion from sample
em_df.rename(columns={'Happy': 'joy', 'Angry': 'anger', 'Surprise': 'surprise', 'Sad': 'sadness', 'Fear': 'fear' }, inplace=True)
```

```python
display(em_df)
```

```python
sel_col=["id","clean_text","primary_emotion"]
t2e_result_df  = pd.merge(sub_df[sel_col], em_df, on='id', how='inner')
display(t2e_result_df)
```
