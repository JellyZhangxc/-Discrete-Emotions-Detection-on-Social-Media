### 2 Load the data and extract a subset as input

```python
import pandas as pd
import nltk
from nltk.tokenize import word_tokenize
from collections import Counter
import re
```

```python
# Ensure you have the necessary nltk data files
nltk.download('punkt')
nrc_lexicon_path='.\Source\NRC-Emotion-Lexicon-Wordlevel-v0.92.txt'
# Load the NRC Emotion Lexicon
nrc_lexicon = pd.read_csv(nrc_lexicon_path, sep='\t', header=None, names=['word', 'emotion', 'association'])

print(len(nrc_lexicon["word"].unique()))

# Filter out the associations
nrc_lexicon = nrc_lexicon[nrc_lexicon['association'] == 1]
display(nrc_lexicon)
```

```python
len(nrc_lexicon["word"].unique())
```

```python
# Filter the DataFrame where association is 1
filtered_df = nrc_lexicon[nrc_lexicon['association'] == 1]

# Count unique words per emotion
unique_counts = filtered_df.groupby('emotion')['word'].nunique()

print(unique_counts)
```

```python
# Count unique words per emotion
filtered_df = filtered_df[(filtered_df['association'] == 1) & (~filtered_df['emotion'].isin(['negative', 'positive']))]
print(len(nrc_lexicon["word"].unique()))
unique_counts = filtered_df.groupby('emotion')['word'].nunique()
print(unique_counts)
```

```python
# Function to get emotions associated with a word
def get_emotions(word):
    #if the word is found, then the emotion(s) get 1 point
    emotions = nrc_lexicon[nrc_lexicon['word'] == word]['emotion'].tolist()
    return emotions

# Function to analyze a sentence
def analyze_sentence(sentence):
    sentence=re.sub(r'[^a-z\s]', '', sentence) ## Remove punctuation and non-alphabetic characters
    words = word_tokenize(sentence.lower())
    emotion_counter = Counter()

    for word in words:
        emotions = get_emotions(word)
        # print(word)
        # print(emotions)
        emotion_counter.update(emotions)
        # print(emotion_counter)

    return emotion_counter
```

### 3  Loop the 900 samples 

```python
#igore trust 
emotions_list = ['joy', 'surprise', 'anger', 'fear', 'sadness', 'disgust', 'anticipation']  # 'neutral',    ->"trust"
em_df = pd.DataFrame(columns=emotions_list+["id"])
display(em_df)
```

```python
for  index, row in sub_df.iterrows():
    sentence=row["text"]
    scores = analyze_sentence(sentence)

    # Create a row with the scores for the specified emotions and 0 for the rest
    em = {emotion: scores.get(emotion, 0) for emotion in emotions_list}
    em['id'] = row['id']

    # Convert the row dictionary to a DataFrame and concatenate with the existing DataFrame
    em_df = pd.concat([em_df, pd.DataFrame([em])], ignore_index=True)
```

```python
#if  all 0 for 'joy', 'surprise', 'anger', 'fear', 'sadness', 'disgust', 'anticipation'  (igore 'trust')  , then neutral
em_df['neutral'] = em_df.apply(lambda row: 1 if all(row[:7] == 0) else 0, axis=1)
#reorder cols
em_df=em_df[['joy', 'surprise', 'anger', 'fear', 'sadness', 'disgust', 'anticipation','neutral','id']]
```

```python
sel_col=["id","text","primary_emotion","Target"]
nrc_result_df  = pd.merge(sub_df[sel_col], em_df, on='id', how='inner')
```
